# 1: 简单的整数划分问题  
总时间限制: 100ms 内存限制: 65536kB  
  
**描述**  
将正整数n 表示成一系列正整数之和，n=n1+n2+…+nk, 其中n1>=n2>=…>=nk>=1 ，k>=1 。  
正整数n 的这种表示称为正整数n 的划分。正整数n 的不同的划分个数称为正整数n 的划分数。  
  
**输入**  
标准的输入包含若干组测试数据。每组测试数据是一个整数N(0 < N <= 50)。  
**输出**  
对于每组测试数据，输出N的划分数。
  
<font color=red>
【注意】<br>
输入有多行且行数不固定。输出有多行，每行一个数表示该行输入的N的划分数。<br>
参考代码：
</font>

```python
while True:
    try: 
        n = int(input())
    except EOFError: 
        break
```

````python
# A two dimensional dynamic programming, conditioning on the smallest component of the division (namely the index j) 

temp = [[0 for _ in range(51)] for __ in range(51)]
temp[1][1] = 1
temp[2][1] = 1
temp[2][2] = 1
for i in range(3,51):
    for j in range(1, i//2 + 1):
        temp[i][j] += sum(temp[i-j][j:])
    temp[i][i] = 1

results = [sum(temp[i]) for i in range(51)]


while True:
    try:
        n = int(input())
        print(results[n])
    except EOFError:
        break
````

# 2: 采药

总时间限制: 1000ms 内存限制: 65536kB

**描述**

辰辰是个很有潜能、天资聪颖的孩子，他的梦想是称为世界上最伟大的医师。为此，他想拜附近最有威望的医师为师。医师为了判断他的资质，给他出了一个难题。医师把他带到个到处都是草药的山洞里对他说：“孩子，这个山洞里有一些不同的草药，采每一株都需要一些时间，每一株也有它自身的价值。我会给你一段时间，在这段时间里，你可以采到一些草药。如果你是一个聪明的孩子，你应该可以让采到的草药的总价值最大。”  
  
如果你是辰辰，你能完成这个任务吗？  

**输入**

输入的第一行有两个整数T（1 <= T <= 1000）和M（1 <= M <= 100），T代表总共能够用来采药的时间，M代表山洞里的草药的数目。接下来的M行每行包括两个在1到100之间（包括1和100）的的整数，分别表示采摘某株草药的时间和这株草药的价值。

**输出**

输出只包括一行，这一行只包含一个整数，表示在规定的时间内，可以采到的草药的最大总价值。

````python
# Recursion not feasible
def optimize(first_n, t_limit):
    global time, value, record
    if record[first_n - 1][t_limit - 1] != None:
        pass
    if min(time[:first_n]) > t_limit:
        record[first_n - 1][t_limit - 1] = 0
    elif first_n == 1:
        record[first_n - 1][t_limit - 1] = value[0]
    else:
        record[first_n - 1][t_limit - 1] = max(optimize(first_n - 1, t_limit), optimize(first_n - 1, t_limit - time[first_n - 1]) + value[first_n - 1])
    return record[first_n - 1][t_limit - 1]
````

````python
# To be optimized
t, m = map(int, input().split())
record = [[0 for _ in range(t+1)] for __ in range(m+1)]
time, value = zip(*([[0, 0]] + [list(map(int, input().split())) for _ in range(m)]))

for i in range(1, m+1): # The i th medicine
    for j in range(1, t+1): #Subject to time limit j
        if min(time[1 : i+1]) > j:
            record[i][j] = 0
        elif i == 1:
            record[i][j] = value[i]
        elif time[i] > j:
            record[i][j] = record[i-1][j]
        else:
            record[i][j] = max(record[i-1][j], record[i-1][j-time[i]]+value[i])

print(record[m][t])
````

# 3: 拦截导弹
总时间限制: 1000ms 内存限制: 65536kB  

**描述**  
某国为了防御敌国的导弹袭击，发展出一种导弹拦截系统。但是这种导弹拦截系统有一个缺陷：虽然它的第一发炮弹能够到达任意的高度，但是以后每一发炮弹都不能高于前一发的高度。某天，雷达捕捉到敌国的导弹来袭。由于该系统还在试用阶段，所以只有一套系统，因此有可能不能拦截所有的导弹。  
  
输入导弹依次飞来的高度（雷达给出的高度数据是不大于30000的正整数），计算这套系统最多能拦截多少导弹。  
  
**输入**  
第一行是一个整数N（不超过15），表示导弹数。  
第二行包含N个整数，为导弹依次飞来的高度（雷达给出的高度数据是不大于30000的正整数）。  
  
**输出**  
一个整数，表示最多能拦截的导弹数。

````python
# Equivalent to finding the longest monotonous decreasing sequence in the input

n = int(input())
height = [0] + list(map(int, input().split()))[::-1]
temp = [0 for _ in range(n+1)]
if n > 0:
    temp[1] = 1

for i in range(2, n+1):
    temp[i] = max([temp[j] for j in range(1, i) if height[j] <= height[i]] + [0]) + 1
    # "+ [0]" is necessary in max([...] + [0]), because [...] may be an empty set

print(max(temp))
````

# 4: 修仙之路
总时间限制: 1000ms 内存限制: 65536kB  
  
**描述**  
修仙之路长漫漫，逆水行舟，不进则退！你过五关斩六将，终于来到了仙界。仙界是一个r行c列的二维格子空间，每个单元格是一个”境界“，每个境界都有等级。你需要任意选择其中一个境界作为起点，从一个境界可以前往上下左右相邻四个境界之一 ，当且仅当新到达的境界等级增加。你苦苦行走，直到所在的境界等级比相邻四个境界的等级都要高为止，一览众山小。请问包括起始境界在内最长修仙路径需要经过的境界数是多少？  
  
**输入**  
第一行为两个正整数，分别为r和c（1<=r,c<=100）。  
接下来有r行，每行有c个0到100000000之间的整数，代表各境界的等级。  
  
**输出**  
输出一行，为最长修仙路径需要经过的境界数（包括起始境界）。

````python
class Maze:
    def __init__(self, r, c, landmap):
        self.row = r
        self.col = c
        self.landmap = landmap
        self.directions = [(1, 0), (0, -1), (-1, 0), (0, 1)]
        self.path_len_record = [[None for _ in range(c)] for __ in range(r)]
    
    def path(self, x, y):
        if self.path_len_record[x][y] != None:
            return self.path_len_record[x][y]
        elif not any([self.can_move(x, y, i, j) for i, j in self.directions]):
            self.path_len_record[x][y] = 1
            return 1
        else:
            fill = max([self.path(x+i, y+j) for i, j in self.directions if self.can_move(x, y, i, j)]) + 1
            # Need not consider the case when [...] in max([...]) is empty, the elif branch has already taken care of it
            self.path_len_record[x][y] = fill
            return fill

    def can_move(self, x, y, i, j):
        try:
            return self.landmap[x+i][y+j] > self.landmap[x][y]
        except:
            return False
        # INDEX ERROR!
    
    def longest_path(self):
        return max([max(lst) for lst in self.path_len_record])

r, c = map(int, input().split())
land = [list(map(int, input().split())) for _ in range(r)]
maze = Maze(r, c, land)
for i in range(r):
    for j in range(c):
        maze.path(i, j)

print(maze.longest_path())
````

# 5: 小木棍
总时间限制: 2000ms 内存限制: 65536kB  
  
**描述**  
小明将一批等长的木棍随机切成最长为50单位的小段。现在他想要将木棍还原成原来的状态，但是却忘记了原来的木棍数量和长度。请写一个程序帮助他计算如果还原成原来的等长木棍，其长度可能的最小值。所有的长度均大于0。  
  
**输入**  
输入包含多个实例。每个实例有两行，第一行是切割后的木棍数量n（最多64个），第二行为n个以空格分开的整数，分别为每根木棍的长度。输入的最后以n为 0 结束。  
  
**输出**  
对于每个实例，输出一行其长度的可能的最小值。

````python
def factor(n):
    return [x for x in range(1, n+1) if n % x == 0]

def can_sum(i, len_lst): 
    ''' 
    Wether i can be represented as the sum of some numbers in len_lst
    If the add up is feasible, return the list of additives so that they can be removed by the test func
    '''
    copy = len_lst[:]
    if i == 0:
        return True, []
    elif copy == []:
        return False, []
    else:
        flag = False
        for ind in range(len(copy)):
            if copy[ind] > i:
                continue
            else:
                copy2 = copy[:]
                copy2.pop(ind)
                flag, lst_of_adds = can_sum(i - copy[ind], copy2)
            if flag:
                return flag, [copy[ind]] + lst_of_adds
        return flag, []

def test(i, len_lst):
    ''' 
    Test whether i is a feasible minimum length using sticks in len_lst
    '''
    copy = len_lst[:]
    if i < copy[0]:
        return False
    elif i == sum(len_lst):
        return True
    elif i == copy[0]:
        copy.pop(0) 
        # Do not use copy.pop(0) as an argument, because the value returned is an element of the list
        return test(i, copy)
    else:
        flag, lst_of_adds = can_sum(i, copy)
        if flag:
            for adds in lst_of_adds:
                copy.pop(copy.index(adds))
            return test(i, copy)
        else:
            return False


while True:
    n = int(input())
    if n == 0:
        break
    length = list(map(int, input().split()))
    length.sort(reverse=True)
    len_sum = sum(length)
    test_lst = factor(len_sum)
    flag = False
    for i in test_lst[:-1]:
        if test(i, length):
            flag = True
            break
    if flag:
        print(i)
    else:
        print(len_sum)
````
